```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
X=np.array([[4,1],[-1,3],[3,5]])
```

```python
mean_val_X=np.mean(X,axis=0)
mean_val_X
```

```python
plt.scatter(X[:,0],X[:,1],label="points")
plt.scatter(mean_val_X[0],mean_val_X[1],label="mean",color="red",marker='^')
plt.show()
```

```python
#EXERCISE1 - 3
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.plot([0,X[0,0]],[0,X[1,0]],[0,X[2,0]],color="blue",label='vector 1')
ax.plot([0,X[0,1]],[0,X[1,1]],[0,X[2,1]],color="red",label='vector 2')
plt.legend()
plt.show()
```

```python
#deviation= y- x_mean*1n
singular_matrix=np.ones((len(X),1))
mean_vectors=mean_val_X.reshape(1,-1).T*singular_matrix.T #2X1*1X3
dev1=X[:,0]-mean_vectors[0]
dev2=X[:,1]-mean_vectors[1]
dev1,dev2,mean_vectors
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.plot([0,mean_vectors[0][0]],[0,mean_vectors[0][1]],[0,mean_vectors[0][2]],linewidth=3)
ax.plot([0,mean_vectors[1][0]],[0,mean_vectors[1][1]],[0,mean_vectors[1][2]],linewidth=1)
ax.plot([0,dev1[0]],[0,dev1[1]],[0,dev1[2]],color="green",label="deviation vector 1")
ax.plot([0,dev2[0]],[0,dev2[1]],[0,dev2[2]],color="blue",label="deviation vector 2")
ax.view_init(elev=30,azim=45)
plt.legend()
plt.show()
```

```python
#cosine similarity
#L1=np.sqrt((dev1**2).sum())
#L2=np.sqrt((dev2**2).sum())
L1=np.linalg.norm(dev1)
L2=np.linalg.norm(dev2)
cos_theta=np.dot(dev1,dev2)/(L1*L2)
theta_degree=np.degrees(np.arccos(cos_theta))
print(f"Angle between deviation vectors is:{theta_degree}")
```

```python
#covariance and correlation matrix
n=len(X)
s11=np.dot(dev1.T,dev1)/n
s22=np.dot(dev2.T,dev2)/n
s12=np.dot(dev1.T,dev2)/n
r12=s12/(np.sqrt(s11*s22))

Sn=np.array([[s11,s12],[s12,s22]])
R=np.array([[1,r12],[r12,1]])
print(f"Covariance matrix:\n{Sn}")
print(f"Correlation matrix:\n{R}")
```

```python
#generalized variance
S=np.array([[252.04,-68.43],
            [-68.43,123.67]])

det_S=np.linalg.det(S)
print(f"Generalized Variance={det_S}")
```

```python
Y=np.array([[1,4,2,5,3],
            [9,12,10,8,11],
            [10,16,12,13,14]]).T

mean_vector=np.mean(Y,axis=0)
D=Y-mean_vector
S=np.cov(D,rowvar=False)
det_S=np.linalg.det(S)
print(f"Generalized Variance={det_S}")
```

```python
import numpy as np

X = np.array([[4, 3, 1],
              [3, 9, 2],
              [1, 2, 1]])

# Compute mean of each column
mean_X = np.mean(X, axis=0)


dev = X - mean_X

# Compute covariance matrix
n = len(X)
Sn = np.dot(dev.T, dev) / n


std_dev = np.sqrt(np.diag(Sn))
R = Sn / np.outer(std_dev, std_dev)


total_sample_variance = np.trace(Sn)


print("Covariance matrix:")
print(Sn)
print("\nCorrelation matrix:")
print(R)
print("\nTotal Sample Variance:", total_sample_variance)

```

**Report on Scatter Plot, Deviation Vectors, Covariance, and Correlation Analysis**

**1. Introduction:**
The purpose of this analysis is to visualize a set of data points, compute deviation vectors from both the mean and the origin, and relate these vectors to covariance and correlation matrices. This helps in understanding the relationships between variables and their statistical properties.

**2. Data Description:**
The dataset consists of three two-dimensional points:
- (9, 1)
- (5, 3)
- (1, 2)

**3. Scatter Plot Analysis:**
A scatter plot was created to visualize the data points in a two-dimensional space. The sample mean of the data was also plotted, marked distinctly to show the central tendency of the data distribution.

**4. Deviation Vectors:**
- **From the Mean Point:** Deviation vectors were drawn from the mean point to each data point. These vectors illustrate how each data point deviates from the average position.
- **From the Origin:** Deviation vectors were also plotted from the origin to observe the data's spread relative to the coordinate system’s origin.

**5. Lengths and Angles of Deviation Vectors:**
The lengths of the deviation vectors were calculated using the Euclidean norm, providing a measure of the magnitude of deviations. The angle between these vectors was computed using the cosine similarity formula, then converted to degrees to understand their directional relationship.

**6. Covariance and Correlation Matrices:**
- **Covariance Matrix (Sn):** This matrix was derived from the dot products of deviation vectors, scaled by the number of data points. It measures how much the variables change together.
- **Correlation Matrix (R):** This matrix normalizes the covariance values to fall between -1 and 1, indicating the strength and direction of linear relationships between variables.

**7. Observations and Relationships:**
- The lengths of the deviation vectors correspond to the square roots of the variances (diagonal elements of the covariance matrix Sn), which reflect the data's spread.
- The cosine of the angle between deviation vectors matches the correlation coefficient (r12) in the correlation matrix R, showing how closely the data points are linearly related.
- The covariance (s12) represents the scaled dot product of the deviation vectors, highlighting the degree of co-movement between variables.
